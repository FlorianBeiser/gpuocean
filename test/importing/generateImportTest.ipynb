# Generate import tests for gpuocean

```python
import os
```

```python
os.environ.get('PYTHONPATH')
```

```python
rootdir = os.path.join('..', '..', 'src', 'gpuocean')
classname = 'ImportTests'
exceptions = ["__init__", "IPythonMagic"]
```

```python


def make_import_test_function(directory, pyfile):
    return """
    def test_import_%s_%s(self):
        %s
        """ % (directory, pyfile, make_import_statement(directory, pyfile))



def make_import_statement(directory, pyfile):
    return "from gpuocean."+directory+" import " + pyfile
    
    
    
def make_test_class_header(classname):
    return """
# WARNING: This file is generated by generateImportTest.ipynb and should not be modified directly. 

import unittest

class %s(unittest.TestCase):

    def setUp(self):
        pass

    def tearDown(self):
        pass

    """ % (classname)

def make_test_class(rootdir, classname, exceptions):
    with open(classname+'.py', 'w') as f:
        f.write(make_test_class_header(classname))

        for d in os.listdir(rootdir):

            pyfiles = [str.replace(f, '.py', '') for f in os.listdir(os.path.join(rootdir, d)) if f.endswith('.py')]
            #print('\n' + d + ':')
            for p in pyfiles:
                if not p in exceptions:
                    #print(p)
                    f.write(make_import_test_function(d, p))
    print('successfully generated test class')

    
```

```python
make_test_class(rootdir, classname, exceptions)
```
